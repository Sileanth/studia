```ocaml
let rec merge cmp xs ys =
  match xs, ys with
  | [], ys -> ys
  | xs, [] -> xs
  | (x :: xs), (y :: ys) when cmp x y -> x :: merge  cmp xs (y :: ys)
  | (x :: xs), (y :: ys) -> y :: merge   cmp (x :: xs) ys
 
```

```ocaml
merge (<) [1 ; 2 ; 3] [2 ; 4 ; 5]
```

```ocaml
let rec merget cmp xs ys zs =
  match xs, ys with
  | [], ys -> List.rev zs @ ys
  | xs, [] -> List.rev zs @ xs
  | (x :: xs), (y :: ys) when cmp x y -> (merget [@tailcall]) cmp xs (y :: ys) (x :: zs)
  | (x :: xs), (y :: ys) -> (merget [@tailcall]) cmp (x :: xs) ys (y :: zs)
```

```ocaml
merget (<) [1 ; 2 ; 3] [2 ; 4 ; 5] []
```

```ocaml
let rec rek xs ys zs =
  match xs with
  | [] -> ys, zs
  | [x] -> x :: ys , zs
  | [x ; y] -> x :: ys , y :: zs
  | x :: y :: rest -> rek rest (x :: ys) ( y :: zs) 

let split xs = rek xs [] []
```

```ocaml
split [1; 2; 2; 3; 4; 5] 
```

```ocaml
let rec sort = function
  | [] -> []
  | [x] -> [x]
  | x :: xs as l -> let (f, s) = split l in merge (<=) (sort f) (sort s)
  
```

```ocaml
sort [1 ; 4; 2; 1; 5; 6; 1; 2]
```

```ocaml
let rec sortet = function
  | [] -> []
  | [x] -> [x]
  | x :: xs as l -> let (f, s) = split l in merget (<=) (sortet f) (sortet s) []
```

```ocaml
sortet [1 ; 4; 2; 1; 5; 6; 1; 2]
```

```ocaml
let rec rand_list = function
  | 0 -> []
  | x -> (Random.int 1000000) :: rand_list (x - 1)
```

```ocaml
let _ = rand_list 10000
```
